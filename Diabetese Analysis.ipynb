```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
#read the data present in the tsv file
df = pd.read_csv("diabetese.tsv",sep="\t")
df.head()
```

```python
df.info()
```

```python
#find mean/average age of diabetic patient
mean_age = df['AGE'].mean()
print(f"Mean age is: {mean_age}")
```

```python
# Variance in the data set
var_age = df['AGE'].var()
print(f"variance in age : {var_age}")
```

```python
#standard Deviation
std_age = df['AGE'].std()
print(f"Standard deviation : {std_age}")
```

```python
#Box ploting for the age feild
plt.figure(figsize=(2,6))
plt.boxplot(df['AGE'],showmeans=True)
plt.grid(color='gray',linestyle='dotted')
plt.xlabel("Age")
plt.ylabel("Years")
plt.show()
```

```python
#Box ploting for bmi
plt.figure(figsize=(6,2))
plt.boxplot(df['BMI'],vert=False,showmeans=True)
plt.grid(color='grey',linestyle='dotted')
plt.xlabel("Value")
plt.ylabel("BMI")
plt.show()
```

```python
#Box plotting for Bp
plt.figure(figsize=(6,2))
plt.boxplot(df['BP'],vert=False,showmeans=True)
plt.grid(color='gray',linestyle='dotted')
plt.xlabel("Value")
plt.ylabel("BP")
plt.show
```

```python
# Making a histogram to know what age of the patients are mostly diabetic

#we can create histogram for the same
df['AGE'].hist(bins=12)
plt.title("Age distribution of the diabetic patient")
plt.xlabel("Age Range")
plt.ylabel("count")
plt.show()
```

```python
#histogram for bmi
df['BMI'].hist(bins=12)
plt.title("Age distribution of the diabetic patient")
plt.xlabel("BMI")
plt.ylabel("Count")
plt.show()
```

```python
#scatter plot in data analysis
hours_studies = [2,3,1,4,2,5,4,6,3,5]
exam_scores = [70,80,60,85,75,90,88,92,80,95]

plt.scatter(hours_studies,exam_scores,color="blue",marker='o')
plt.xlabel("Hours Studied")
plt.ylabel("Exam Score")
plt.grid(color="gray",linestyle="dotted")
plt.title("Scatter plot : Hours studied vs exam scores")
plt.show()
```

```python
#scatter plot of bmi and disease progression
plt.scatter(df["BMI"],df['Y'],color="blue")
plt.xlabel("BMI")
plt.ylabel("Disease Progression")
plt.grid(color="gray",linestyle="dotted")
plt.title("BMI V/s Disease Progression")
plt.show()
```

```python
#scatter plot of blood type and disease progression
plt.scatter(df["S5"],df['Y'],color="blue")
plt.xlabel("S5")
plt.ylabel("Disease Progression")
plt.grid(color="gray",linestyle="dotted")
plt.title("S5 V/s Disease Progression")
plt.show()
```
